```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import plotly.express as px

# -----------------------
# 1. Sample Dataset
# -----------------------
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May"],
    "Sales": [200, 250, 300, 280, 350],
    "Expenses": [150, 180, 220, 200, 260],
    "Customers": [50, 60, 70, 65, 80],
    "Products": ["A", "B", "C", "D", "E"],
    "Revenue": [5000, 7000, 6000, 8000, 9000]
}
df = pd.DataFrame(data)

# -----------------------
# 2. Comparison (Bar/Column)
# -----------------------
plt.figure(figsize=(12,6))
plt.bar(df["Products"], df["Revenue"], color="skyblue")
plt.title("Bar Chart → Revenue by Product")
plt.xlabel("Products")
plt.ylabel("Revenue")
plt.show()

# -----------------------
# 3. Trend (Line Chart)
# -----------------------
plt.plot(df["Month"], df["Sales"], marker="o", color="green")
plt.title("Line Chart → Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()

# -----------------------
# 4. Proportion (Pie/Doughnut)
# -----------------------
plt.pie(df["Revenue"], labels=df["Products"], autopct="%1.1f%%", startangle=140)
plt.title("Pie Chart → Revenue Share")
plt.show()

# Doughnut (Pie + White Circle)
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(df["Revenue"], labels=df["Products"], autopct="%1.1f%%")
centre_circle = plt.Circle((0,0),0.70,fc="white")
fig.gca().add_artist(centre_circle)
plt.title("Doughnut Chart → Revenue Share")
plt.show()

# -----------------------
# 5. Distribution (Histogram & Box)
# -----------------------
plt.hist(df["Revenue"], bins=5, color="purple", alpha=0.7)
plt.title("Histogram → Revenue Distribution")
plt.show()

plt.boxplot(df["Revenue"], patch_artist=True)
plt.title("Box Plot → Revenue Spread")
plt.show()

# -----------------------
# 6. Relationship (Scatter/Bubble)
# -----------------------
plt.scatter(df["Sales"], df["Expenses"], color="red")
plt.title("Scatter Plot → Sales vs Expenses")
plt.xlabel("Sales")
plt.ylabel("Expenses")
plt.show()

plt.scatter(df["Sales"], df["Expenses"], s=df["Customers"]*10, alpha=0.5, color="orange")
plt.title("Bubble Chart → Sales vs Expenses (Bubble = Customers)")
plt.xlabel("Sales")
plt.ylabel("Expenses")
plt.show()

# -----------------------
# 7. Heatmap (Advanced Insights)
# -----------------------
corr = df[["Sales","Expenses","Revenue","Customers"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heatmap → Correlation Between Variables")
plt.show()

# -----------------------
# 8. Funnel Chart (Plotly)
# -----------------------
funnel_data = {
    "Stage": ["Visitors", "Leads", "Qualified Leads", "Proposals", "Customers"],
    "Count": [1000, 600, 400, 200, 100]
}
funnel_df = pd.DataFrame(funnel_data)
fig = px.funnel(funnel_df, x="Count", y="Stage", title="Funnel Chart → Sales Pipeline")
fig.show()
```
