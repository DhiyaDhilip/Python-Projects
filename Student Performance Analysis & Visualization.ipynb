```python
# ðŸ”¹ Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ðŸ”¹ Load Dataset
data = pd.read_csv(r"D:/Data Sets/student_performance.csv")
print(data.head())
print("Columns:", data.columns.tolist())
print("Shape:", data.shape)
data.dtypes
data.info()
data.describe()
```

```python
data.isnull().sum()
```

```python
# Histogram â€“ Self Study Hours
plt.hist(data['weekly_self_study_hours'], bins=5, color='skyblue', edgecolor='black')
plt.title("Weekly Self Study Hours")
plt.xlabel("Hours")
plt.ylabel("Count")
plt.show()
```

```python
# Boxplot â€“ Total Score by Grade
sns.boxplot(x='grade', y='total_score', data=data, palette='Set2')
plt.title("Total Score by Grade")
plt.xlabel("Grade")
plt.ylabel("Score")
plt.show()
```

```python
# Bar Chart â€“ Average Score by Grade
grade_avg = data.groupby('grade')['total_score'].mean()
plt.bar(grade_avg.index, grade_avg.values, color='orange', edgecolor='black')
plt.title("Average Score by Grade")
plt.xlabel("Grade")
plt.ylabel("Average Score")
plt.show()
```

```python
# Pie Chart â€“ Grade Distribution
grade_counts = data['grade'].value_counts()
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Grade Distribution")
plt.show()
```

```python
# Scatter Plot â€“ Attendance vs Total Score
plt.scatter(data['attendance_percentage'], data['total_score'], color='purple', alpha=0.6)
plt.title("Attendance vs Total Score")
plt.xlabel("Attendance (%)")
plt.ylabel("Total Score")
plt.show()
```

```python
# Heatmap â€“ Correlation
corr_cols = data[['weekly_self_study_hours', 'attendance_percentage', 'class_participation', 'total_score']]
sns.heatmap(corr_cols.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
